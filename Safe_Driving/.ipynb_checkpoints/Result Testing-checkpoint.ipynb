# For Real Life Static Images

```python
cascPathface = os.path.dirname(cv2.__file__) + "/data/haarcascade_frontalface_default.xml"
cascPatheyes = os.path.dirname(cv2.__file__) + "/data/haarcascade_eye.xml"

faceCascade = cv2.CascadeClassifier(cascPathface)
eyeCascade = cv2.CascadeClassifier(cascPatheyes)
```

```python
face_array = []
eye_array = []
path = os.path.join("./TestingViolaJones")

for image in os.listdir(path):
    img = cv2.imread(os.path.join(path,image))
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    faces = faceCascade.detectMultiScale(gray,scaleFactor=1.1, minNeighbors=5,minSize=(60, 60),flags=cv2.CASCADE_SCALE_IMAGE)
    x = faces[0][0]
    y = faces[0][1]
    w = faces[0][2]
    h = faces[0][3]
    cv2.rectangle(gray, (x, y), (x + w, y + h),(0,255,0), 2)
    faceROI = gray[y:y+h,x:x+w]
    face_array.append(faceROI)
    eyes = eyeCascade.detectMultiScale(faceROI)
    
    for (x2,y2,w2,h2) in eyes:
        eyeROI = faceROI[y2:y2+h2,x2:x2+w2]
        eye_array.append(eyeROI)
        break
```

```python
img = cv2.imread("./TestingViolaJones/6.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

faces = faceCascade.detectMultiScale(gray,scaleFactor=1.1, minNeighbors=5,minSize=(60, 60),flags=cv2.CASCADE_SCALE_IMAGE)
x = faces[0][0]
y = faces[0][1]
w = faces[0][2]
h = faces[0][3]
cv2.rectangle(gray, (x, y), (x + w, y + h),(0,255,0), 2)
faceROI = gray[y:y+h,x:x+w]
eyes = eyeCascade.detectMultiScale(faceROI)

for (x2,y2,w2,h2) in eyes:
    eyeROI = faceROI[y2:y2+h2,x2:x2+w2]
    cv2.rectangle(faceROI, (x2, y2), (x2 + w2, y2 + h2),(0,255,0), 2)
    break
```

```python
plt.imshow(gray)
plt.show()
```

## Face Prediction

```python
classification = ['Anger', 'Closed_Eyes', 'Happy', 'Neutral', 'Open_Eyes', 'Yawn']

for face in face_array:
    plt.imshow(face)
    plt.show()
    resized_face = resize(face, (224, 224, 3))
    final_face = np.expand_dims(resized_face,axis = 0)
    pred_face = Xception.predict(final_face)

    list_index = [0,1,2,3,4,5]
    for i in range(6):
        for j in range(6):
            if pred_face[0][list_index[i]] > pred_face[0][list_index[j]]:
                temp = list_index[i]
                list_index[i] = list_index[j]
                list_index[j] = temp
    for i in range(6):
        print(classification[list_index[i]], ':', pred_face[0][list_index[i]]*100, '%')
   
```

## Eye Prediction

```python
for eye in eye_array:
    plt.imshow(eye)
    plt.show()
    resized_eye = resize(eye, (224, 224, 3))
    final_eye = np.expand_dims(resized_eye,axis = 0)
    pred_eye = Xception.predict(final_eye)

    list_index = [0,1,2,3,4,5]
    for i in range(6):
        for j in range(6):
            if pred_eye[0][list_index[i]] > pred_eye[0][list_index[j]]:
                temp = list_index[i]
                list_index[i] = list_index[j]
                list_index[j] = temp
    for i in range(6):
        print(classification[list_index[i]], ':', pred_eye[0][list_index[i]]*100, '%')
```
